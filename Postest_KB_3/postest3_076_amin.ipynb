```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
df = pd.read_csv('players_15.csv')
```

```python
pd.set_option('display.max_columns', None)
df.head()
```

```python
data = df.groupby('overall').mean()
data.head()
```

## Menampilkan Data Power Shoot Berdasarkan Overall menggunakan Scatter plot

```python
plt.figure(figsize=(15,10))
plt.scatter(df['overall'], df['shooting'])
plt.title('Data power shoot berdasarkan overall')
plt.xlabel('Overall')
plt.ylabel('Shooting Power')
plt.show()
```

## Menampilkan Presentasi kaki yang disukai dengan menggunakan Pie Chart

```python
data = df.preferred_foot.value_counts()
plt.pie(data, labels=data.index, autopct="%1.1f%%")
plt.title('Persentase Kaki')
plt.show()
```

## Menampilkan semua data dengan menggunakan HeatMap

```python
plt.figure(figsize=(25,10))
cor_matrix = df.corr()
sns.heatmap(cor_matrix, cmap='coolwarm', annot=True, center=0)
plt.show()
```

## Menampilkan data pace dengan Histogram

```python
plt.hist(df['pace'], bins=25)
sns.set_style('darkgrid')
plt.title('Histogram data pace')
plt.ylabel('Frekuensi Pace')
plt.xlabel('Pace')
plt.show()
```
