## Menggunakan fungsi pandas yaitu read_csv untuk membaa dataset yang akan di kelola, disini saya menggunakan dataset "players_15.csv", untuk mengelola datanya.

## Di dalam kolom tadi ada beberapa attribute yang memiliki beberapa kolom, jadi disini akan Menampilkan informasi jumlah record, jumlah attribute, nama attribute, jumlah attribute dengan tipe angka. Masukkan informasi tersebut dalam dictionary dengan key sesuai nama informasinya. dan akan menampilkan isi dari dictionary tersebut sebagus mungkin.

```python
num_records = len(data)
num_attributes = len(data.columns)
attribute_names = list(data.columns)
numeric_attributes = len(data.select_dtypes(include=['number']).columns)
info_dict = {
    'Jumlah Record'                   : num_records,
    'Jumlah Atribut'                  : num_attributes,
    'Nama Atribut'                    : attribute_names,
    'Jumlah Atribut dengan Tipe Angka': numeric_attributes
}
for key, value in info_dict.items():
    print(f'{key}: {value}')
```

##  Menampilkan nilai dataset pada baris dimulai dari NIM saya yaitu 76 hingga baris terakhir, dan mulai dari kolom 4 hingga  kolom atau attribute terakhir, dengan menggunakan metode slice "loc"

```python
data.loc[76:, data.columns[3:]]
```

##  Menampilkan nilai dataset pada baris dimulai dari NIM saya yaitu 76 hingga baris terakhir, dan mulai dari kolom 4 hingga  kolom atau attribute terakhir, dengan menggunakan metode slice "iloc"
